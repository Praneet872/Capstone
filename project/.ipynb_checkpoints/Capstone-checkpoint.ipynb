```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('FPA_FOD_20170508.sqlite')

# Create a cursor object
cursor = conn.cursor()

# Execute a SQL query to fetch the column names from the table
cursor.execute("PRAGMA table_info(your_table)")

# Fetch all rows
columns_info = cursor.fetchall()

# Extract column names
column_names = [column[1] for column in columns_info]

# Construct a SELECT query using the column names
select_query = "SELECT * FROM your_table"

# Execute the SELECT query
cursor.execute(select_query)

# Fetch all rows
rows = cursor.fetchall()

# Print column names
print("Column Names:")
print(column_names)

# Print the data
print("Data:")
for row in rows:
    print(row)

# Close the connection
conn.close()
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('FPA_FOD_20170508.sqlite')

# Create a cursor object
cursor = conn.cursor()

# Execute a SQL query to fetch table names from the database schema
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")

# Fetch all rows
tables = cursor.fetchall()

# Print the table names
print("Tables:")
for table in tables:
    print(table[0])

# Close the connection
conn.close()
```

```python
# Connect to the SQLite database
conn = sqlite3.connect('FPA_FOD_20170508.sqlite')

# Create a cursor object
cursor = conn.cursor()

# List of tables
tables = [
    'spatial_ref_sys',
    'spatialite_history',
    'sqlite_sequence',
    'geometry_columns',
    'spatial_ref_sys_aux',
    'views_geometry_columns',
    'virts_geometry_columns',
    'geometry_columns_statistics',
    'views_geometry_columns_statistics',
    'virts_geometry_columns_statistics',
    'geometry_columns_field_infos',
    'views_geometry_columns_field_infos',
    'virts_geometry_columns_field_infos',
    'geometry_columns_time',
    'geometry_columns_auth',
    'views_geometry_columns_auth',
    'virts_geometry_columns_auth',
    'sql_statements_log',
    'SpatialIndex',
    'ElementaryGeometries',
    'KNN',
    'Fires',
    'idx_Fires_Shape',
    'idx_Fires_Shape_node',
    'idx_Fires_Shape_rowid',
    'idx_Fires_Shape_parent',
    'NWCG_UnitIDActive_20170109'
]

# Iterate through each table
for table in tables:
    print("Table:", table)
    
    # Execute a SQL query to select all data from the current table
    cursor.execute("SELECT * FROM {}".format(table))
    
    # Fetch all rows
    rows = cursor.fetchall()
    
    # Print the data
    for row in rows:
        print(row)
    
    print()  # Add a newline between tables

# Close the connection
conn.close()
```
